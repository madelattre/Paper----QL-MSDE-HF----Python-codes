## Data processing

We use the neural data provided as an example in several `R` packages, including `mixedsde`. The dataset consists of `N=240` trajectories of membrane potentials, each containing `n=2000` points spaced by a time step of `h=0.00015` seconds. This dataset has been the subject of several published analyses, in which the authors primarily employed Ornstein-Uhlenbeck processes with Gaussian random effects and fixed effects in the drift, along with a constant diffusion coefficient.

```python
conversion.set_conversion(pandas2ri.converter)

r('load("examples/neuronal.data.rda")')
r_obj = r['neuronal.data']
y_init = list(r_obj[0])
time = list(r_obj[1])
```

Before the analysis, we scale the signal by multiplying the membrane potential measurements by a factor of `200`. The aim is to make visible an inter-trajectory variability that would otherwise be challenging to discern.

```python
y = jax.numpy.array(y_init) * 200
times = jax.numpy.array(time)
h = 0.00015

nb_trajectories = y.shape[0]
time_mat = jax.numpy.tile(times, (nb_trajectories, 1))
```

The following graph shows the dynamics observed in the data.

```python
plt.figure(figsize=(10, 6))
for i in range(y.shape[0]):
    plt.plot(
        time_mat[0,:], jax.numpy.array(y[i, :]), color="black", linewidth=1.5
    )
    
plt.xlabel("Time")
plt.ylabel("Values")
plt.title("Data Plot")
plt.show()
```

## Model Definitions

Here we define the models used to analyze the neural data:

- **(N1)**  
$Y_i(t) = \tau_i \int_0^t \left(\frac{\varphi_{i1} Y_i(s)}{\sqrt{1+Y_i(s)^2}} + \frac{\varphi_{i2}}{\sqrt{1+Y_i(s)^2}}\right) ds + \sqrt{\tau_i} \int_0^t \exp(\eta \, \mathrm{atan}(Y_i(s))) \, dW_i(s)$

- **(N2)**  
$Y_i(t) = \tau_i \int_0^t \left(\frac{\varphi_{i1} Y_i(s)}{\sqrt{1+Y_i(s)^2}} + \frac{\varphi_{i2}}{\sqrt{1+Y_i(s)^2}}\right) ds + \sqrt{\tau_i} \int_0^t \exp(\eta s / 2) \, dW_i(s)$

- **(N3)**  
$Y_i(t) = \tau_i \int_0^t (\varphi_{i1} + \varphi_{i2} Y_i(s)) \, ds + \int_0^t \sqrt{\tau_i (1 + \eta Y_i(s)^2)} \, dW_i(s)$

Model (N3) is a more general formulation that includes the Ornstein–Uhlenbeck process as a special case when $\eta=0$.



```python
## Drift functions

def a_n1_2(y, time):
    a_value = jax.numpy.array(
        [y  / jax.numpy.sqrt(1 + y * y), 1.0 /
         jax.numpy.sqrt(1 + y * y) + 0.0 * time]
    )
    return a_value

def a_n3(y, time):
    a_value = jax.numpy.array(
        [y , 1.0 + 0.0 * time]
    )
    return a_value


## Diffusion coefficients

def c_n1(y, eta, time):
    c_value = jax.numpy.exp(jax.numpy.arctan(y) * eta) + 0.0 * time
    return c_value


def c_n2(y, eta, time):
    c_value = jax.numpy.exp(eta * time / 2) 
    return c_value

def c_n3(y, eta, time):
    c_value = jax.numpy.sqrt(1 + eta * y * y) + 0.0 * time
    return c_value
```

The lines below define a `Mixedsde` object for each model, including the model structure, the covariance of the drift random effects, and the initial values used for parameter estimation.

```python
model_name = 'n3' # replace by 'n1', 'n2' or 'n3' to reproduce the analysis for another model
```

```python

eta = jax.numpy.array([0.0])

theta_tau = {
    'weibull_loc': {'loc': 'N/A', 'shape': 'N/A', 'scale': 'N/A'}, 
}

theta_phi = {
    'normal': {'mu': jax.numpy.array([0.0,0.0]), 'omega2': jax.numpy.array([[0.0,0.0],[0.0,0.0]])} 
} 

if (model_name == 'n1'):
    model = Mixedsde(drift_func=a_n1_2,diffusion_func=c_n1,drift_re_param=theta_phi['normal'],diffusion_fixed_effect=eta, diffusion_re_dist='weibull_loc',diffusion_re_param=theta_tau['weibull_loc'])
    
    covariance_to_estimate=jax.numpy.array([[0.0,0.0],[0.0,1.0]],dtype=int)

    init = {
        'diffusion': {
            'eta': jax.numpy.array([0.0]), 
            'theta_tau':  {'shape': 3.0, 'scale': 3.0, 'loc': 3.0}
        },
        'drift': {
            'mu': jax.numpy.array([-5.0,10.0]), 
            'omega2': jax.numpy.array([[0.0,0.0],[0.0,1.0]])
            } 
    } 

if (model_name == 'n2'):
    model = Mixedsde(drift_func=a_n1_2,diffusion_func=c_n2,drift_re_param=theta_phi['normal'],diffusion_fixed_effect=eta, diffusion_re_dist='weibull_loc',diffusion_re_param=theta_tau['weibull_loc'])
    
    covariance_to_estimate=jax.numpy.array([[0.0,0.0],[0.0,1.0]],dtype=int)

    init = {
        'diffusion': {
            'eta': jax.numpy.array([0.0]), 
            'theta_tau':  {'shape': 3.0, 'scale': 3.0, 'loc': 3.0}
        },
        'drift': {
            'mu': jax.numpy.array([-5.0,10.0]), 
            'omega2': jax.numpy.array([[0.0,0.0],[0.0,1.0]])
            } 
    } 
    
if (model_name == 'n3'):
    model = Mixedsde(drift_func=a_n3,diffusion_func=c_n3,drift_re_param=theta_phi['normal'],diffusion_fixed_effect=eta, diffusion_re_dist='weibull_loc',diffusion_re_param=theta_tau['weibull_loc'])
    
    covariance_to_estimate=jax.numpy.array([[1.0, 1.0],[1.0, 1.0]],dtype=int)

    init = {
        'diffusion': {
            'eta': jax.numpy.array([0.0]), 
            'theta_tau':  {'shape': 3.0, 'scale': 3.0, 'loc': 3.0}
        },
        'drift': {
            'mu': jax.numpy.array([-5.0,10.0]), 
            'omega2': jax.numpy.array([[1.0,0.0],[0.0,1.0]])
            } 
    } 
     
```

## Estimation

```python
eta_hat, theta_tau_hat, tau_hat, mu_hat, omega2_hat = model.fit_mixed_sde(y,time_mat,init,covariance_to_estimate,method='joint')
```

We obtain the following parameter estimations:

```python
print("Fixed effect in the diffusion coefficient", eta_hat.item())
print("Distribution of the random effects in the diffusion coefficient", theta_tau_hat)
print("Mean of the random effects in the drift", mu_hat)
print("Variance of the random effects in the drift")
print(omega2_hat)
```

## Model adequacy

### Distribution of the random effect in the diffusion coefficient

```python
x_vals = jax.numpy.linspace(min(tau_hat) - 0.5, max(tau_hat) + 0.5, 300) 

def logpdf_weibull_ind(x, k, lam, loc):
                return (
                    (k - 1) * jax.numpy.log((x - loc) / lam)
                    - ((x - loc) / lam) ** k
                    + jax.numpy.log(k / lam)
                )

            
pdf_vals = jax.numpy.exp(logpdf_weibull_ind(x_vals, k=theta_tau_hat[0], lam=theta_tau_hat[1], loc=theta_tau_hat[2]))

plt.figure(figsize=(7,5))
plt.hist(tau_hat[:,-1], bins=30, density=True, alpha=0.6, color="skyblue", label="Data (normalized histogram)")
plt.plot(x_vals, pdf_vals, 'r-', lw=2, label="Fitted Weibull PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.title("Histogram and Fitted Weibull Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


```

```python
loc, shape, scale = weibull_min.fit(tau_hat)
weibull_transformed_tau_hat = (tau_hat[:,-1]-loc)/scale
probplot(weibull_transformed_tau_hat, dist="weibull_min", sparams=(shape,), plot=plt)
plt.title("QQ-Plot for the adjusted distribution")
plt.show()
```

### Visual predictive checks

```python

if (model_name == 'n1'):
    def drift(y, phi, tau):
        a_value = phi[0] * y  / jax.numpy.sqrt(1 + y * y) + phi[1] / jax.numpy.sqrt(1 + y * y) 
        return tau * a_value

    def diffusion(y, tau, eta, t):
        c_value = np.sqrt(tau) * jax.numpy.exp(jax.numpy.arctan(y) * eta)
        return c_value

if (model_name == 'n2'):
    def drift(y, phi, tau):
        a_value = phi[0] * y  / jax.numpy.sqrt(1 + y * y) + phi[1] / jax.numpy.sqrt(1 + y * y) 
        return tau * a_value

    def diffusion(y, tau, eta, t):
        c_value = np.sqrt(tau) * jax.numpy.exp(eta * t / 2) + 0.0 * y
        return c_value
    
if (model_name == 'n3'):
    def drift(y, phi, tau):
        a_value =  phi[0] * y  + phi[1]
        return tau * a_value

    def diffusion(y, tau, eta, t):
        c_value = np.sqrt(tau)*np.sqrt(1 + eta * y * y) 
        return c_value
    
def simu(random, fixed, h, n, y0):
    y = np.zeros(n + 1)
    y[0] = y0
    phi = random[0:2]
    tau = random[2]
    eta = fixed[0]
   

    for j in range(n):
         
        y[j+1] = y[j] + drift(y[j], phi, tau) * h + diffusion(y[j],
                                                                    tau, eta, h*(j+1)) * np.random.normal(scale=np.sqrt(h))

    return y


def simu_joint(eta, theta_tau, mu, omega2, N, h, n, y0):
   
    shape, scale, loc = theta_tau 

    tau = (weibull_min.rvs(c=shape, size=N)) * scale  + loc 
    phi = np.random.multivariate_normal(mu, omega2, N)
    re = np.column_stack((phi, tau))
    
    y = np.zeros((N, n + 1))
    for i in range(N):
        y[i,:] = simu(random=re[i,:], fixed=eta, h=h, n=n, y0=y0) 

    return {'y': y, 'tau': tau, 'phi': phi}
```
